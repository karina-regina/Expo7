```python
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge
from scipy.ndimage import rotate, shift
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
from sklearn.linear_model import LinearRegression
import scipy.stats as sp
import pandas as pd
import statsmodels.api as sm
import math
import cmath

#set up the figure 
plt.ion()
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_xlim(-100, 100)
ax.set_ylim(-100, 100)
plt.title('Stupid cone')

# Define the center of the cone, you can change the max radius to whatever deems fit 
#this just defines the circle area where the point can be randomized
radius = 90
angle = 2 * np.pi * np.random.rand()
r = radius * np.sqrt(np.random.rand())
 
# while True:
#     x_c = round(r * np.cos(angle))
#     y_c = round(r * np.sin(angle))
    
#     if x_c > 10 and y_c>10:
#         x_c = x_c
#         y_c = y_c

#these are the randomized center points of the plume (it is not yet done that it cannot be less than 10)
x_c = round(r * np.cos(angle))
y_c = round(r * np.sin(angle))
center=[x_c,y_c]
print(center)

# Generate a random radius within the range [10, 100]
random_radius = np.random.randint(10, 100)
print(random_radius)

#Generate random angle between 30-60 degrees for the cone but making sure that it is all downward 
#its for the wedge function
while True:
    m_a = np.random.randint(240, 255)
    m_b = np.random.randint(285, 300)

    if m_b - 270 == 270 - m_a:
        min_a = m_a
        max_a = m_b
        break


# Create a Wedge to represe
# nt the cone pointing downwards
cone = Wedge(center, random_radius, min_a, max_a, width=random_radius, edgecolor='b', facecolor='blue')
ax.add_artist(cone)

#these are all the possible radius with difference of 5.625m 
radius= [10, 15.625, 21.25, 26.875, 32.5, 38.125, 43.75, 49.375, 55, 60.625, 66.25, 71.875, 77.5, 83.125,88.75,94.375,100]

#divided into the phases 
# phase_1 = [55]
# phase_2 = [32.5, 77.5]
# phase_3 = [21.25, 43.75, 66.25, 88.75]
# phase_4 = [15.625, 26.875, 38.125, 49.375, 60.625, 71.875,83.125,94.375]
phase_1 = [55.5]
phase_2 = [33.25, 77.75]
phase_3 = [22.125, 44.375, 66.625, 88.875]
phase_4 = [16.5625, 27.6875, 38.8125, 49.9375, 61.0625, 72.1875,83.125,94.375]

#plotting all the phases
for i in radius:
    if i in [10,100]:
        ax.add_artist(plt.Circle((0, 0), i, edgecolor='black', facecolor='none'))
    elif i in phase_1:
        ax.add_artist(plt.Circle((0, 0), i, edgecolor='r', facecolor='none'))
    elif i in phase_2:
        ax.add_artist(plt.Circle((0, 0), i, edgecolor='g', facecolor='none'))
    elif i in phase_3:
        ax.add_artist(plt.Circle((0, 0), i, edgecolor='b', facecolor='none'))
    elif i in phase_4:
        ax.add_artist(plt.Circle((0, 0), i, edgecolor='m', facecolor='none', linestyle= "-."))

plt.show()
```
