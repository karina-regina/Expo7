```python
def coordinates(phase):
    """
    arguments: phase which is a list of tuples (tuple has two elements, radius and sample numbers)

    output: 2 lists, x and y coordinates of the phase given 

    """
    xcor = []
    ycor = []
    for j in phase: 
        r = j[0]
        n = j[1]
        angle = 360/n
        a = 0 
        for i in range(n):
            #get x and y coordinate of angle 
            x = round(r * m.cos(m.radians(a)), 3)
            y = round(r * m.sin(m.radians(a)), 3)

            #put coordinates in list 
            xcor.append(x)
            ycor.append(y)

            #go to next angle 
            a += angle 
            
            #double check the angle doesnt exceed 360 
            if a >= 360:
                break
    return xcor, ycor


phase_1 = [(55, 12)]
phase_2 = [(33.25, 7), (77.75, 17)]
phase_3 = [(22.125, 5),(44.375, 10), (66.625, 14), (88.875, 19)]
phase_4 = [(16.5625, 4), (27.6875, 6), (38.8125, 8), (49.9375, 11), (61.0625, 13), (72.1875, 16), (93.3125, 18), (97.4375, 20)]

small_big = [(55, 12), (33.25, 7), (77.75, 17),(22.125, 5),(88.875, 19), (44.375, 10), (66.625, 14),(16.5625, 4),(97.4375, 20),(27.6875, 6),(93.3125, 18), (38.8125, 8),(72.1875, 16),(49.9375, 11), (61.0625, 13) ]
all_phases = [(55, 12),(33.25, 7), (77.75, 17),(22.125, 5),(44.375, 10), (66.625, 14), (88.875, 19),(16.5625, 4), (27.6875, 6), (38.8125, 8), (49.9375, 11), (61.0625, 13), (72.1875, 16), (93.3125, 18), (97.4375, 20)]
coordinates(small_big)
coordinates(all_phases)


```
